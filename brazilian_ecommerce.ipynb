```python
import pandas as pd
import os

dataset_folder = 'brazilian-ecommerce'

# 1. Load Datasets
orders = pd.read_csv(os.path.join(dataset_folder, 'olist_orders_dataset.csv'))
items = pd.read_csv(os.path.join(dataset_folder, 'olist_order_items_dataset.csv'))
products = pd.read_csv(os.path.join(dataset_folder, 'olist_products_dataset.csv'))
translations = pd.read_csv(os.path.join(dataset_folder, 'product_category_name_translation.csv'))
payments = pd.read_csv('brazilian-ecommerce/olist_order_payments_dataset.csv')

print("Datasets loaded successfully!")
```

```python
# 2. End-to-End Merging
# Merging orders table with items table and then adding product table
df = orders.merge(items, on='order_id', how='left')
df = df.merge(products, on='product_id', how='left')
df = df.merge(translations, on='product_category_name', how='left')
df = df.merge(payments, on='order_id', how='left')
```

```python
df.head(5)
```

```python
# 3. Data Cleaning
# Handle missing product names by using the original category name if translation is missing.
df['product_category_name_english'] = df['product_category_name_english'].fillna(df['product_category_name'])
```

```python
df.head()
```

```python
# Standardize date formats (Crucial for SQL time-series analysis)
date_columns = ['order_purchase_timestamp', 'order_delivered_customer_date', 'order_estimated_delivery_date']
for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
# Feature Engineering: Calculate Delivery Performance
df['delivery_days'] = (df['order_delivered_customer_date'] - df['order_purchase_timestamp']).dt.days
```

```python
# Handle outliers in Price using IQR method
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['price'] >= (Q1 - 1.5 * IQR)) & (df['price'] <= (Q3 + 1.5 * IQR))]
```
